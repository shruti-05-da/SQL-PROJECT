```python
from sqlalchemy import create_engine

def get_engine():
    return create_engine(
        'mysql+mysqlconnector://root:shruti0509@localhost:3306/sql_projects',
        connect_args={
            "connection_timeout": 60,
            "use_pure": True
        }
    )

customers.to_sql('customers', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
location.to_sql('location', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
order_items.to_sql('order_items', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
payment.to_sql('payment', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
review.to_sql('review', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
order.to_sql('orders', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
product.to_sql('product', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
seller.to_sql('seller', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
prod_category.to_sql('prod_category', con=get_engine(), index=False, if_exists='replace', chunksize=1000)
 
```
